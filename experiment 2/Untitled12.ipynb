```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def visualize_time_series(data, title="Time Series Plot", xlabel="Time", ylabel="Value"):
    """
    Visualizes time series data using a line plot.

    Args:
        data (list, numpy.ndarray, or pandas.Series): The time series data.
        title (str): The title of the plot.
        xlabel (str): The label for the x-axis.
        ylabel (str): The label for the y-axis.
    """

    if isinstance(data, list):
        data = np.array(data)

    if isinstance(data, np.ndarray):
      time_points = np.arange(len(data))
      plt.plot(time_points, data)
    elif isinstance(data, pd.Series):
        plt.plot(data.index, data.values)
    else:
        raise TypeError("Input data must be a list, numpy.ndarray or pandas.Series")

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()


# Example usage with a NumPy array:
data = np.random.rand(50)
visualize_time_series(data)

# Example usage with a pandas Series:
dates = pd.date_range(start='2024-01-01', periods=10)
values = np.random.randint(0, 100, size=10)
time_series = pd.Series(values, index=dates)

visualize_time_series(time_series, title="Pandas Series Plot", xlabel="Date", ylabel="Value")

# Example with a list
my_list = [1, 5, 2, 8, 3]
visualize_time_series(my_list)
```

```python
# prompt: Implement programs for visualizing time series data. give 5 types of visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Date' column is already in datetime format
# If not, convert it using pd.to_datetime(df['Date'])

def visualize_time_series_data(df):
    # 1. Line Plot
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['Sales'])  # Replace 'Sales' with your relevant column
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales Over Time (Line Plot)')
    plt.grid(True)
    plt.show()

    # 2. Scatter Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Date'], df['Sales'])  # Replace 'Sales' with your relevant column
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales Over Time (Scatter Plot)')
    plt.grid(True)
    plt.show()

    # 3. Area Plot
    plt.figure(figsize=(10, 6))
    plt.fill_between(df['Date'], df['Sales'], alpha=0.5) # Replace 'Sales' with relevant column
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales Over Time (Area Plot)')
    plt.grid(True)
    plt.show()

    # 4. Bar Chart (for discrete time intervals)
    # Group data by month/week/year and plot sales
    df['Month'] = df['Date'].dt.to_period('M')
    monthly_sales = df.groupby('Month')['Sales'].sum()
    plt.figure(figsize=(10, 6))
    monthly_sales.plot(kind='bar')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.title('Monthly Sales (Bar Chart)')
    plt.grid(axis='y') #grid only on y-axis
    plt.show()

    # 5. Seasonal Decomposition (requires statsmodels)
    # This decomposes time series into trend, seasonality, and residuals
    from statsmodels.tsa.seasonal import seasonal_decompose
    result = seasonal_decompose(df['Sales'], model='additive') # or 'multiplicative'
    result.plot()
    plt.show()


visualize_time_series_data(df)
```
